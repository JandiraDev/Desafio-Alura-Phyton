### Importaci√≥n de datos



```python
import pandas as pd

url = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_1%20.csv"
url2 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_2.csv"
url3 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_3.csv"
url4 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_4.csv"

tienda = pd.read_csv(url)
tienda2 = pd.read_csv(url2)
tienda3 = pd.read_csv(url3)
tienda4 = pd.read_csv(url4)

tienda.head()
```

#1. An√°lisis de facturaci√≥n



```python
# Calcular el ingreso total de cada tienda sumando la columna 'Precio'
ingreso_tienda_1 = tienda['Precio'].sum()
ingreso_tienda_2 = tienda2['Precio'].sum()
ingreso_tienda_3 = tienda3['Precio'].sum()
ingreso_tienda_4 = tienda4['Precio'].sum()

# Mostrar los resultados
print(f"Ingreso total Tienda 1: {ingreso_tienda_1}")
print(f"Ingreso total Tienda 2: {ingreso_tienda_2}")
print(f"Ingreso total Tienda 3: {ingreso_tienda_3}")
print(f"Ingreso total Tienda 4: {ingreso_tienda_4}")
```

# 2. Ventas por categor√≠a

```python
# Cantidad de productos vendidos por categor√≠a en cada tienda

print("üõí Tienda 1 - Ventas por categor√≠a:")
print(tienda['Categor√≠a del Producto'].value_counts())
print("\n-----------------------------")

print("üõí Tienda 2 - Ventas por categor√≠a:")
print(tienda2['Categor√≠a del Producto'].value_counts())
print("\n-----------------------------")

print("üõí Tienda 3 - Ventas por categor√≠a:")
print(tienda3['Categor√≠a del Producto'].value_counts())
print("\n-----------------------------")

print("üõí Tienda 4 - Ventas por categor√≠a:")
print(tienda4['Categor√≠a del Producto'].value_counts())

```

# 3. Calificaci√≥n promedio de la tienda


```python
# Calcular calificaci√≥n promedio por tienda
calificacion_prom_tienda_1 = tienda['Calificaci√≥n'].mean()
calificacion_prom_tienda_2 = tienda2['Calificaci√≥n'].mean()
calificacion_prom_tienda_3 = tienda3['Calificaci√≥n'].mean()
calificacion_prom_tienda_4 = tienda4['Calificaci√≥n'].mean()

# Mostrar resultados
print(f"‚≠ê Calificaci√≥n promedio Tienda 1: {calificacion_prom_tienda_1:.2f}")
print(f"‚≠ê Calificaci√≥n promedio Tienda 2: {calificacion_prom_tienda_2:.2f}")
print(f"‚≠ê Calificaci√≥n promedio Tienda 3: {calificacion_prom_tienda_3:.2f}")
print(f"‚≠ê Calificaci√≥n promedio Tienda 4: {calificacion_prom_tienda_4:.2f}")
```

# 4. Productos m√°s y menos vendidos

```python
# Productos m√°s vendidos y menos vendidos en cada tienda

def mostrar_mas_y_menos_vendidos(df, nombre_tienda):
    print(f"\nüìä {nombre_tienda}")
    productos = df['Producto'].value_counts()
    print("üîù M√°s vendido:")
    print(productos.head(1))
    print("üîª Menos vendido:")
    print(productos.tail(1))

mostrar_mas_y_menos_vendidos(tienda, "Tienda 1")
mostrar_mas_y_menos_vendidos(tienda2, "Tienda 2")
mostrar_mas_y_menos_vendidos(tienda3, "Tienda 3")
mostrar_mas_y_menos_vendidos(tienda4, "Tienda 4")
```

# 5. Env√≠o promedio por tienda

```python
# Calcular el costo de env√≠o promedio por tienda
envio_prom_tienda_1 = tienda['Costo de env√≠o'].mean()
envio_prom_tienda_2 = tienda2['Costo de env√≠o'].mean()
envio_prom_tienda_3 = tienda3['Costo de env√≠o'].mean()
envio_prom_tienda_4 = tienda4['Costo de env√≠o'].mean()

# Mostrar resultados
print(f"üì¶ Costo de env√≠o promedio Tienda 1: {envio_prom_tienda_1:.2f}")
print(f"üì¶ Costo de env√≠o promedio Tienda 2: {envio_prom_tienda_2:.2f}")
print(f"üì¶ Costo de env√≠o promedio Tienda 3: {envio_prom_tienda_3:.2f}")
print(f"üì¶ Costo de env√≠o promedio Tienda 4: {envio_prom_tienda_4:.2f}")
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

url = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_1%20.csv"
url2 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_2.csv"
url3 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_3.csv"
url4 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_4.csv"

tienda = pd.read_csv(url)
tienda2 = pd.read_csv(url2)
tienda3 = pd.read_csv(url3)
tienda4 = pd.read_csv(url4)

tienda.head()

# Crear figura con subplots para las 4 tiendas
fig, axes = plt.subplots(2, 2, figsize=(14, 10))
fig.suptitle("Distribuci√≥n geogr√°fica de ventas por tienda", fontsize=16)

# Lista de tiendas y sus respectivos ejes
tiendas = [(tienda, "Tienda 1"), (tienda2, "Tienda 2"), (tienda3, "Tienda 3"), (tienda4, "Tienda 4")]
axes_flat = axes.flatten()

# Crear gr√°fico de dispersi√≥n para cada tienda
for i, (df, nombre) in enumerate(tiendas):
    sns.scatterplot(
        x=df['lon'],
        y=df['lat'],
        alpha=0.4,
        s=40,
        ax=axes_flat[i]
    )
    axes_flat[i].set_title(nombre)
    axes_flat[i].set_xlabel("Longitud")
    axes_flat[i].set_ylabel("Latitud")

plt.tight_layout(rect=[0, 0, 1, 0.96])
plt.show()
```
